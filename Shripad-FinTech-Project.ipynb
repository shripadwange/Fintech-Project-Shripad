```python
fig, ax= plt.subplots()

ax.scatter(df['longitude'],df['latitude'])
ax.set_title('longitude-latitude')
ax.set_xlabel('longitude')
ax.set_ylabel('latitude')

plt.show()
```

```python
fig, ax= plt.subplots()

ax.scatter(df['total_rooms'],df['total_bedrooms'])
ax.set_title('Rooms and Bedrooms')
ax.set_xlabel('Total Rooms')
ax.set_ylabel('Bedrooms')

plt.show()
```

```python
fig, ax= plt.subplots()

ax.scatter(df['median_house_value'],df['median_income'])
ax.set_title('Scatter plot of House value')
ax.set_xlabel('Price of house')
ax.set_ylabel('Income')

plt.show()
```

```python
df['ocean_proximity'].value_counts()
```

```python
colors={'<1H OCEAN':'r','INLAND':'b','NEAR OCEAN':'y','NEAR BAY':'w','ISLAND':'g'}
fig, ax= plt.subplots()
for i in range(len(df['median_house_value'])):
    ax.scatter(df['median_house_value'][i],df['median_income'][i],color=colors[df['ocean_proximity'][i]])
    
ax.set_title('House value scatter ploting')
ax.set_xlabel('Price of House')
ax.set_ylabel('Income')

plt.show()
```

```python
columns=df.columns.drop(['ocean_proximity'])
x_data=range(0,df.shape[0])
fig,ax=plt.subplots()
for c in columns:
    ax.plot(x_data,df[c])
ax.set_title('Housing Data')
ax.legend(labels=[])
plt.show()
```
